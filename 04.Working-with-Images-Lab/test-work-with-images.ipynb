```python
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import binary_opening
from scipy.ndimage.filters import convolve
```

```python
hamburger = imread('test-data/hamburger.jpg')
ham_red = hamburger[:,:,0]
ham_green = hamburger[:,:,1]
ham_blue = hamburger[:,:,2]
```

```python
plt.hist(ham_red.ravel(), bins = 256, color = "red")
plt.hist(ham_green.ravel(), bins = 256, color = "green")
plt.hist(ham_blue.ravel(), bins = 256, color = "blue")

plt.show()
```

```python
print(f'hamburger average red: {np.average(ham_red)}')
print(f'hamburger average green: {np.average(ham_green)}')
print(f'hamburger average blue: {np.average(ham_blue)}')
```

```python
ham_blue_binary = ham_blue > 36.79
structuring_element = np.ones((3, 3), dtype=bool)
opened_image = binary_opening(ham_blue_binary, structure=structuring_element)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(ham_blue_binary, cmap='gray')
ax[0].set_title('Original Blue Channel (Binary)')

ax[1].imshow(opened_image, cmap='gray')
ax[1].set_title('After Binary Opening')

plt.show()
```
